```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
```

```python
dset_path = './synth_dset_grin_Nmodes=113_degen=True_len=5000_mirr=36.mat'
dset = loadmat(dset_path)
```

```python
print(dset.keys())
print(dset['fields'].shape)
```

```python
idx = np.random.randint(0, dset['intens'].shape[-1])

pls = []
fig, axs = plt.subplots(1, 3, figsize=(18,4))
pls.append(axs[0].imshow(np.square(np.abs(dset['fields'][..., idx]))))
pls.append(axs[1].imshow(dset['intens'][..., idx]))
pls.append(axs[2].imshow(dset['intens'][..., idx] - np.square(np.abs(dset['fields'][..., idx]))))

_ = [plt.colorbar(pls[i], ax=axs[i]) for i in range(len(axs))]
axs[0].set_title(r"$|y_{i}|^{2}$")
axs[1].set_title(r"$|y_{i_{sim}}|^{2}$")
axs[2].set_title(r"$|y_{i_{sim}}|^{2} - |y_{i}|^{2}$")
# _ = [axs[i].set_xticks([]) for i in range(len(axs))]
# _ = [axs[i].set_yticks([]) for i in range(len(axs))]
```
